### (3) Perform the cleaning and pre-processing

Before starting the modeling tasks, you need to clean your dataset which includes manipulating those null values and reduce the dataframe to the essential columns. You can either drop rows, fill with zeros or other techniques you consider necessary.

In this task you can drop the following columns:
`['avland','avtot', 'exland', 'extot','avland2', 'avtot2', 'exland2', 'extot2']`

```python
# I trying to cut off the irrealtvent values first.
data02 = data.drop(columns=['avland','avtot', 'exland', 'extot','avland2', 'avtot2', 'exland2', 'extot2'],axis=1)
data02.head()
```

```python
# If I keep the NAN,the regressor wiil failed
data03 = data02.fillna(0)
data03
```

### (4) Define target variable ($y$) and predictors ($X$) and split data into training and test data
    Since we are predicting the market value, 'fullval' would be your target column. Feel free to use the columns you find more useful.

```python
# because I finish the previous corr analysis par, so I can choose the high colored X 
input_var = ['ltdepth','stories','blddepth','community_board','bldfront','council_district','census_tract','ltfront','excd1','block','boro']
target_var = 'fullval'
```

```python
dataset = data03.copy()
X_train, X_test, y_train, y_test = train_test_split(data03[input_var], data03[target_var], test_size=0.25, random_state=42)
```

# 3. Create OLS model:
### (1) Show train and test results 
    Please use R-squared and Mean Squared Error (MSE) to evaluate your model

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
from sklearn.metrics import mean_absolute_percentage_error
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
```

### (2) Show regression coefficients

```python
%%time
#LinearRegressor 
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train, y_train)
y_test_pred = model.predict(X_test)
dataset.loc[X_test.index, 'fullval_pred'] = y_test_pred

r2 = r2_score(y_test, y_test_pred)
MAPE = mean_absolute_percentage_error(y_test, y_test_pred)
MAE = mean_absolute_error(y_test, y_test_pred)
print('r2_score: '+str(round(r2, 3)))
print('mean_absolute_percentage_error: '+str(round(MAPE, 3)))
print('mean_absolute_error: '+str(round(MAE, 3)))

dataset[['fullval','fullval_pred']].plot(figsize=(10,6))
```

```python
%%time
#LGBMRegressor model
print('model: LGBMRegressor')
model = lgb.LGBMRegressor(n_estimators=100) 
model.fit(X_train, y_train)
y_test_pred = model.predict(X_test)
dataset.loc[X_test.index, 'fullval_pred'] = y_test_pred

r2 = r2_score(y_test, y_test_pred)
MAPE = mean_absolute_percentage_error(y_test, y_test_pred)
MAE = mean_absolute_error(y_test, y_test_pred)
print('r2_score: '+str(round(r2, 3)))
print('mean_absolute_percentage_error: '+str(round(MAPE, 3)))
print('mean_absolute_error: '+str(round(MAE, 3)))

dataset[['fullval','fullval_pred']].plot(figsize=(10,6))
```

```python
#Showing the impact of each item x on the regression model can help delete some low-impact parameters to optimize the model
lgb.plot_importance(model, ax=None, height=0.2, xlim=None, ylim=None,importance_type = 'auto')
```

```python
%%time
##Âª∫Á´ãRamdomForestRegressor model  ÈÄüÂ∫¶ÊÖ¢‰ΩÜÊïàÊûúÂá∫‰ºó
from sklearn.ensemble import RandomForestRegressor
model = RandomForestRegressor()
model.fit(X_train, y_train)
y_test_pred = model.predict(X_test)
dataset.loc[X_test.index, 'fullval_pred'] = y_test_pred

r2 = r2_score(y_test, y_test_pred)
MAPE = mean_absolute_percentage_error(y_test, y_test_pred)
MAE = mean_absolute_error(y_test, y_test_pred)
print('r2_score: '+str(round(r2, 3)))
print('mean_absolute_percentage_error: '+str(round(MAPE, 3)))
print('mean_absolute_error: '+str(round(MAE, 3)))

dataset[['fullval','fullval_pred']].plot(figsize=(10,6))
```
